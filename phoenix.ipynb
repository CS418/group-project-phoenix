```python
# The dataset is in text format, so converting that to dataframe for easier manipulation
def convert_text_to_df(filename):
    
    with open(filename) as file:
        lines = file.readlines()
    
    file.close()   
    df = pd.DataFrame(columns = ['Sentence', 'Emotion'])

    for line in lines:
        x = line.split(';')
        df = df.append({'Sentence' : x[0], 'Emotion' : x[1].strip()},
                       ignore_index = True)
    
    return df

df_train = convert_text_to_df('train.txt')
df_test = convert_text_to_df('test.txt')
df_val = convert_text_to_df('val.txt')

# Concat all data together for easier preprocessing
df = pd.concat([df_train, df_test, df_val])
```

```python
# Create a word cloud draw function
def word_cloud(wc, df):
    plt.figure(figsize=(10,10))
    plt.title(df+' word cloud', size=16)
    plt.imshow(wc)
    plt.axis('off')

emotion_label_list = ['anger', 'fear', 'joy', 'love', 'sadness']
complete_text_input = ""

for emotion in emotion_label_list:
    text_input = ' '.join(
       [text for text in df.loc[df['Emotion'] == emotion,'Sentence']])
    wc = WordCloud(width=500, height=500).generate(text_input)
    word_cloud(wc, emotion)
    complete_text_input += text_input

    
```

```python
# Function to generate ngrams and output c most common ngrams
def generate_common_N_grams(filt_sentence,ngram=1,c=2):
    temp=zip(*[filt_sentence[i:] for i in range(0,ngram)])
    ans=[' '.join(ngram) for ngram in temp]
    common_bigrams = ['dont know','dont want','cant help',
        'even though','little bit','every day','every time','cant bothered']
    a = [i for i in ans if i not in common_bigrams]
    return Counter(a).most_common(c)
```

```python
for emotion in emotion_label_list:
    text_input_em = []
    text_input_em = (' '.join(
        [text for text in df.loc[df['Emotion'] == emotion,'Sentence']]))
    stop_words = set(stopwords.words('english'))
    
    # Eliminating few more unrelevant words. 
    stop_words.add('feel')
    stop_words.add('feeling')
    stop_words.add('like')
    stop_words.add('im')
    stop_words.add('really')
    stop_words.add('get')
    stop_words.add('http')
    stop_words.add('href')
    stop_words.add('www')
    
    word_tokens = word_tokenize(text_input_em)
    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
    frequency = Counter(filtered_sentence).most_common(15)
    frequent_bigrams = generate_common_N_grams(filtered_sentence,2,5)
    plt.figure(figsize=(15, 5))
    
    # Visualizing most common words
    a_str = 'Frequency of 15 most common words for class: ' + emotion
    b_str = 'Frequency of 5 most common bigrams for class: ' + emotion
    plt.title(a_str)
    plt.bar(*zip(*frequency))
    plt.show()
    
    plt.figure(figsize=(15, 5))
    # Visualizing most common bigrams
    plt.title(b_str)
    plt.bar(*zip(*frequent_bigrams))
    plt.show()
```

```python
# Data Cleaning continues
stopwords = set(stopwords.words('english'))
stemmer = PorterStemmer()

def count_words(user_input):
    num_words = 0
    tokens = nltk.tokenize.word_tokenize(user_input)
    for word in tokens:
         if word not in string.punctuation:
            num_words += 1
    return num_words

def get_tokens(inp_str):
    inp_str = inp_str.lower()
    return inp_str.split()

def remove_numbers(input_str):
    return re.sub("[^a-zA-Z]", " ", input_str)

def pre_process(dataframe, column):
    
    tokens_processed = []    
    for sentence in dataframe[column]:
        text_inp = remove_numbers(sentence)
        words = get_tokens(sentence)
        words = [stemmer.stem(w) for w in words if w not in stopwords 
                 and w not in string.punctuation]
        words = " ".join(words)
        tokens_processed.append(words)
        
    return tokens_processed
    

token = RegexpTokenizer(r'[a-zA-Z0-9]+')
vectorizer = CountVectorizer(stop_words='english', max_df=0.5, 
                    min_df=3, ngram_range=(1,1),tokenizer = token.tokenize)

df['Processed_Sentence'] = pre_process(df, 'Sentence')
X = vectorizer.fit_transform(df.Processed_Sentence)
Y = df.Emotion.values
print(Y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,
                                                    random_state=42)
```

# Clustering - Task 3

```python
# The dataset is in text format, so converting that to dataframe for easier manipulation
def extract_data_for_clustering(filename):
    
    sentence_list = []
    with open(filename) as file:
        lines = file.readlines()
    
    file.close()   
    
    for line in lines:
        x = line.split(';')
        # Extracting only sentences without emotion label
        sentence_list.append(x[0])
    
    return sentence_list

train_sentences = extract_data_for_clustering('train.txt')
test_sentences = extract_data_for_clustering('test.txt')
val_sentences = extract_data_for_clustering('val.txt')

# Concat all data together for easier preprocessing
data_sentences = train_sentences + test_sentences + val_sentences
```

```python
len(data_sentences) #20000 sentences
```

```python
data_sentences[0] # sample
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer_2 = TfidfVectorizer(stop_words='english')
X_data = vectorizer_2.fit_transform(data_sentences)
```

```python
X_data
```

```python
from sklearn.cluster import KMeans

model_8 = KMeans(n_clusters=5, init='k-means++', max_iter=100, n_init=1)
model_8.fit(X_data)
```

```python
model_8_cluster_centre = model_8.cluster_centers_.argsort()[:, ::-1]

cluster_1_top_10 = model_8_cluster_centre[0, :10]
cluster_2_top_10 = model_8_cluster_centre[1, :10]
cluster_3_top_10 = model_8_cluster_centre[2, :10]
cluster_4_top_10 = model_8_cluster_centre[3, :10]
cluster_5_top_10 = model_8_cluster_centre[4, :10]

cluster_top_10 = []
cluster_top_10.append(cluster_1_top_10)
cluster_top_10.append(cluster_2_top_10)
cluster_top_10.append(cluster_3_top_10)
cluster_top_10.append(cluster_4_top_10)
cluster_top_10.append(cluster_5_top_10)
```
